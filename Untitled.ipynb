```python
import torch
import pickle
from torch.distributions.beta import Beta
import numpy as np
import matplotlib.pyplot as plt
```

```python
x0 = 0.66
y0 = 0.9
alpha = (2*y0+1)*x0/(3*(y0-x0))
beta = (1/x0-1)*alpha
print(alpha, beta)
grades = Beta(alpha, beta)
x = np.arange(0,1,0.01)
n = x.shape[0]
y = np.zeros(n)
for i in range(n):
    y[i] = 1-grades.sample()
y.sort()
median = y[n//2]
mean = y.mean()
print(n,1-mean, 1-median)
fig = plt.plot(x,y)
```
